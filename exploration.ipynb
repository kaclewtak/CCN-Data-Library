```python
import pandas as pd
from ydata_profiling import ProfileReport
from pathlib import Path
import os
```

```python
# Load data for cores_files

base_path = "../../data/primary_studies"
cores_files = []

# Go through all directories to find files ending _cores.csv
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith("_cores.csv"):
            full_path = os.path.join(root, file)
            cores_files.append(full_path)
            print(f"Found: {file}")

print(f"\nTotal: {len(cores_files)} files")
```

```python
# Generate report for all cores_files found
for i, file in enumerate(cores_files, 1):
    df = pd.read_csv(file)

    # Get filename stem
    file_name = Path(file).stem

    # Generate the report
    profile = ProfileReport(df, title=f"Profiling Report - {file_name}")

    # Save to HTML
    profile.to_file(f"EDA/Kacper/Cores_ydata_reports/ydata_report_{file_name}.html")

    print(f"\nReport {i}/len{cores_files} generated: ydata_report_{file_name}")
```
