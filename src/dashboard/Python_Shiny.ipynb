# Sample notebook to run the CCN Dashboard

This provides a simple, single cell setup and run approach. We import the dashboard application, nest_asyncio (for running in an interactive system), and uvicorn to run the dashboard as a locally served.application

```python
import nest_asyncio
import shiny_dashboard
import uvicorn

# Load the dashboard app
app = shiny_dashboard.app

# Apply nest_asyncio to allow the dashboard to run in an interactive environment (Jupyter Notebook).
nest_asyncio.apply()

# Runs the unified dashboard with Data Explorer as the starting tab.
# Import a dataset from the embedded spreadsheet in Data Explorer, then use
# Satellite Search and QA Dashboard with the same session dataset.
config = uvicorn.Config(app, port=8000, host="127.0.0.1", log_level="info")
server = uvicorn.Server(config)
await server.serve()
```
